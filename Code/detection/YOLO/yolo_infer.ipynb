```python
import os
from ultralytics import YOLO
import cv2
```

```python
# video_path = "/home/theo/Documents/Unif/Master/ChimpRec/Code/Superpixel R-CNN/Video/Video_1.mp4"
# video_path = "/home/theo/Documents/Unif/Master/ChimpRec-Datasets/ChimpBehave/chimpbehave/original/Colebe4_241_325.mp4"

# video_path = "/home/theo/Documents/Unif/Master/ChimpRec-Datasets/Videos chimprec/penda7.mp4"
video_path = "/home/theo/Documents/Unif/Master/ChimpRec-Datasets/Videos chimprec/Chimpanzees - Camera trap footage.mp4"
video_path_out = 'out.mp4'
```

```python
cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))


model_version = "train9"
model_path = f"/home/theo/Documents/Unif/Master/ChimpRec/Code/detection/YOLO/runs/detect/{model_version}/weights/best.pt"

# Load a model
model = YOLO(model_path)  # load a custom model

threshold = 0.5

while ret:

    results = model(frame)[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyAllWindows()
```
